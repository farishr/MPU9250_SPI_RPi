```python
# Fourier Transform on Accelerometer and Gyroscope Data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plotting Time Series Data of No DLPF
# df = pd.read_csv("Acc_Data_NoLPF_v2.csv")
# df = pd.read_csv("Acc_Data_LPF6.csv")
file_name = "Acc_Gyr_Data_LPF7.csv"
df = pd.read_csv(file_name)
print(df)
N = 5000
Ax = df["Ax"][0:N]
Ay = df["Ay"][0:N]
Az = df["Az"][0:N]
Gx = df["Gx"][0:N]
Gy = df["Gy"][0:N]
Gz = df["Gz"][0:N]
X = np.arange(0, N, 1)
```

```python
fig, axis = plt.subplots(6,1)
axis[0].plot(X, Ax)
axis[1].plot(X, Ay)
axis[2].plot(X, Az)
axis[3].plot(X, Gx)
axis[4].plot(X, Gy)
axis[5].plot(X, Gz)
# axis[0].scatter(X, Ax, color="green")
# axis[1].scatter(X, Ay, color="red")
# axis[2].scatter(X, Az)
axis[0].set_title("Ax", fontweight ="bold", fontsize = 18)
axis[1].set_title("Ay", fontweight ="bold", fontsize = 18)
axis[2].set_title("Az", fontweight ="bold", fontsize = 18)
axis[3].set_title("Gx", fontweight ="bold", fontsize = 18)
axis[4].set_title("Gy", fontweight ="bold", fontsize = 18)
axis[5].set_title("Gz", fontweight ="bold", fontsize = 18)

for x in range(6):
    axis[x].set_xlabel('Time')
    axis[x].set_ylabel('Amplitude')

fig.set_figwidth(30)
fig.set_figheight(40)
plt.savefig(file_name[:-4] + "_Time_Series.png")
plt.show()
```

```python
# Applying Fourier Tranform
from scipy.fft import rfft, rfftfreq

Ax_FT = rfft(Ax)
Ay_FT = rfft(Ay)
Az_FT = rfft(Az)
Gx_FT = rfft(Gx)
Gy_FT = rfft(Gy)
Gz_FT = rfft(Gz)
f = rfftfreq(N, 1/20000)

fig2, axis2 = plt.subplots(6,1)
axis2[0].plot(f[1:], abs(Ax_FT[1:]))
axis2[1].plot(f[1:], abs(Ay_FT[1:]))
axis2[2].plot(f[1:], abs(Az_FT[1:]))
axis2[3].plot(f[1:], abs(Gx_FT[1:]))
axis2[4].plot(f[1:], abs(Gy_FT[1:]))
axis2[5].plot(f[1:], abs(Gz_FT[1:]))
# axis2[0].scatter(f, Ax_FT, color="green")
# axis2[1].scatter(f, Ay_FT, color="red")
# axis2[2].scatter(f, Az_FT)
axis2[0].set_title("Ax_FT", fontweight ="bold", fontsize = 18)
axis2[1].set_title("Ay_FT", fontweight ="bold", fontsize = 18)
axis2[2].set_title("Az_FT", fontweight ="bold", fontsize = 18)
axis2[3].set_title("Gx_FT", fontweight ="bold", fontsize = 18)
axis2[4].set_title("Gy_FT", fontweight ="bold", fontsize = 18)
axis2[5].set_title("Gz_FT", fontweight ="bold", fontsize = 18 )

for x in range(6):
    axis2[x].set_xlabel('Frequency')
    axis2[x].set_ylabel('Power')

fig2.set_figwidth(30)
fig2.set_figheight(40)
plt.savefig(file_name[:-4] + "_Fourier_Transform.png")
plt.show()
```
